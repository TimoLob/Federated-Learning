```python
from manim import *
from manim_ml.neural_network import Convolutional2DLayer, FeedForwardLayer, NeuralNetwork, MaxPooling2DLayer, ImageLayer
import random
```

```python
class Smartphone(Group):
    def __init__(self,width=1,height = None,color=BLUE,**kwargs):
        Mobject.__init__(self,**kwargs)
        aspect_ratio = 9/17
        if height == None:
            height = width * aspect_ratio
        self.frame = RoundedRectangle(height=width,width=height,color=color,corner_radius=0.3)
        self.add(self.frame)
        self.button = Circle(width * 0.07,color=color)
        self.add(self.button)
        self.button.shift(DOWN * height * 0.75)
        self.nn = NeuralNetwork([
            FeedForwardLayer(4),
            FeedForwardLayer(3),
            FeedForwardLayer(2),
        ],
        layer_spacing = 0.25                       
        )
        self.nn.rotate(1.5*PI)
        self.add(self.nn)
```

```python
class Server(Group):
    def __init__(self,width=1,height = None,color=BLUE,**kwargs):
        Mobject.__init__(self,**kwargs)
        aspect_ratio = 1
        if height == None:
            height = width * aspect_ratio
        self.frame = RoundedRectangle(height=width,width=height,color=color,corner_radius=0.3)
        self.add(self.frame)
        self.nn = NeuralNetwork([
            FeedForwardLayer(4),
            FeedForwardLayer(3),
            FeedForwardLayer(2),
        ],
        layer_spacing = 0.25,
        )
        self.nn.rotate(1.5*PI)
        self.add(self.nn)
```

```python
%%manim -qh HFL

class HFL(Scene):
    def construct(self):
        # Initial Setup
        num_clients = 6
        clients = [
            Smartphone(width=2.7).shift(LEFT*5 + UP * 2).shift(RIGHT * i * 2) for i in range(num_clients)
        ]
        self.add(*clients)
        server = Server(width = 2, color = RED).shift(DOWN*2)
        self.add(server)
        label = Text("Round 0",font_size = 20).shift(LEFT*6+UP*3.7)
        self.add(label)
        # Training round
        iterations = 5
        K = int(num_clients/2) # Number of clients per round
        epochs_per_iteration = 2
        for i in range(iterations):
            self.play(Transform(label,Text(f"Round {i}",font_size = 20).shift(LEFT*6+UP*3.7)))
            random_clients = random.sample(clients,K)
            global_model_copies = [
                server.nn.copy() for _ in random_clients
            ]
            animations = [
                gm.animate.move_to(c) for gm,c in zip(global_model_copies,random_clients)
            ]
            self.play(animations)
            forward_passes = [
                client.nn.make_forward_pass_animation(run_time=1) for client in random_clients
            ]
            for i in range(epochs_per_iteration):
                self.play(forward_passes)
            animations = [
                gm.animate.next_to(server,UP,buff=0.1) for gm in global_model_copies
            ]
            self.play(animations)
            animations = [
                gm.animate.move_to(server) for gm in global_model_copies
            ]
            self.play(animations)
```
